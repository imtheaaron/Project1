```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
path = os.path.join('complete_data.csv')
df = pd.read_csv(path)
df.head()
```

```python
df.Average_price = df.Average_price.round(1)
df.head()
```

```python
price_df = pd.DataFrame({
    'City': df.City,
    'Price': df.Average_price
})
price_df = price_df.sort_values(by=['Price'], ascending=False)
price_df.head()
```

```python
sns.set()
price_df.plot(kind='bar', figsize=(14,10), x=price_df.City)
plt.show()
```

```python
sun_df = pd.DataFrame({
    'City': df.City,
    'Sun': df.SunHours
})
sun_df = sun_df.sort_values(by=['Sun'], ascending=False)
sun_df.head()
```

```python
sun_df.plot(kind='bar', figsize=(14,10), x=sun_df.City, color='orange')
plt.ylim(1000,1400)
plt.show()
```

```python
temp_df = pd.DataFrame({
    'City': df.City,
    'Temp': df.MaxTempSum
})
temp_df = temp_df.sort_values(by=['Temp'], ascending=False)
temp_df.head()
```

```python
temp_df.plot(kind='bar', figsize=(14,10), x=temp_df.City, color='red')
plt.ylim(200,350)
plt.show()
```

```python
df = df.sort_values(by = ['Average_price'], ascending= False)
df.head()
```

```python
price_rank = list(range(3,198,3))
df["Price_Rank"] = price_rank

df.head()
```

```python
df = df.sort_values(by = ['SunHours'])
df.head()
```

```python
sun_hrs_rank = list(range(2,132,2))
df["SunHours_Rank"] = sun_hrs_rank

df.head()
```

```python
df = df.sort_values(by = ['MaxTempSum'], ascending= False)
df.head()
```

```python
temp_rank = list(range(1,66))
df["Max_Temp_Rank"] = temp_rank

df.head()
```

```python
df["Index"] = list(range(1,66))
```

```python
df = df.set_index('Index')
df.head()
```

```python
df["Overall_Rank"] = df.SunHours_Rank + df. Max_Temp_Rank + df.Price_Rank
df.head()
```

```python
df = df.sort_values(by = ['Overall_Rank'], ascending= False)
df.head()
```
