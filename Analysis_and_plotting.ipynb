```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
import numpy as np
```

```python
path = os.path.join('complete_data.csv')
df = pd.read_csv(path)
df.head()
```

```python
df.Average_price = df.Average_price.round(1)
df.head()
```

```python
price_df = pd.DataFrame({
    'City': df.City,
    'Price': df.Average_price
})
price_df = price_df.sort_values(by=['Price'], ascending=False)
price_df.head()
```

```python
sns.set()
price_df.plot(kind='bar', figsize=(14,10), x=price_df.City)
plt.show()
```

```python
sun_df = pd.DataFrame({
    'City': df.City,
    'Sun': df.SunHours
})
sun_df = sun_df.sort_values(by=['Sun'], ascending=False)
sun_df.head()
```

```python
sun_df.plot(kind='bar', figsize=(14,10), x=sun_df.City, color='orange')
plt.ylim(1000,1400)
plt.show()
```

```python
temp_df = pd.DataFrame({
    'City': df.City,
    'Temp': df.MaxTempSum
})
temp_df = temp_df.sort_values(by=['Temp'], ascending=False)
temp_df.head()
```

```python
temp_df.plot(kind='bar', figsize=(14,10), x=temp_df.City, color='red')
plt.ylim(200,350)
plt.show()
```

```python
df = df.sort_values(by = ['SunHours'])
df.head()
```

```python
sun_hrs_rank = list(range(2,132,2))
df["SunHours_Rank"] = sun_hrs_rank

df.head()
```

```python
df = df.sort_values(by = ['MaxTempSum'], ascending= False)
df.head()
```

```python
temp_rank = list(range(1,66))
df["Max_Temp_Rank"] = temp_rank

df.head()
```

```python
#now, let's try a different ranking method for the different cities - rather than forced ranking,
#let's normalize the max for each category to 100 (or some other value) and get results based on that.
#for price and temp, we will also have to adjust to make the highest values the lowest on our narmalized scale.
df.Average_price.max()
df['Normal_Price'] = 101 - (df.Average_price / df.Average_price.max()) * 100
df.head()
```

```python
#looks good!
df.tail()
```

```python
df['Normal_Sun'] = (df.SunHours / df.SunHours.max()) * 100
df.head()
```

```python
df.tail()
```

```python
#since temperature is not as important, we will normailze this to 50 for the max.  However, here we want a high score for COOL
#locations, so we need to adjust the data so places with higher max temps rank lower.

df['Normal_Temp'] =  76 - (50 * (df.MaxTempSum / df.MaxTempSum.max()))
```

```python
df.tail()
```

```python
df['Final_Rank'] = df.Normal_Temp + df.Normal_Sun + df.Normal_Price
df.head()
```

```python
#sort by these new rankings and inspect the results.
df = df.sort_values(by = ['Final_Rank'], ascending= False)
df
```

```python
#let's now graph these new results:

xtickloc = np.arange(65)
plt.figure(figsize=(15,10))
p1 = plt.bar(xtickloc, df.Normal_Price, color='darkgreen', label='Normalized Price Rank')
p2 = plt.bar(xtickloc, df.Normal_Sun, bottom=df.Normal_Price, color='orange', label='Normalized Sun Hours Rank')
p3 = plt.bar(xtickloc, df.Normal_Temp, bottom=(df.Normal_Price + df.Normal_Sun), color='crimson', label='Normalized Temperature Rank')
plt.xticks(xtickloc, df.City, rotation='vertical')
plt.title('Ranking of Cities for Solar Installations - Normalized Scores', size=17)
plt.ylabel('Overall Score', size=15)
plt.xlabel('City', size=15)
plt.legend()
plt.show()
```
