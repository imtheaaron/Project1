```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib
```

```python
land_data = os.path.join("county ppsf sales data csv.csv")
```

```python
land_df = pd.read_csv(land_data)
land_df.head()
```

```python
# Are we missing any data for given cities? - No
land_df.count()
```

```python
land_df.columns
```

```python
grouped_land_data = land_df.groupby(["Region"])
grouped_land_data.head(1)
```

```python
# This is the mean of the Avg Median.PPSF
avg_price = grouped_land_data.mean()
avg_price.head()
```

```python
avg_price = avg_price.reset_index()
avg_price.head()
```

```python
#rename the average price column
df = avg_price.rename(columns={'Avg. Median Ppsf': 'Average_price'})
df.head()
```

```python
#now we will add a new column that contains that state name for all cities (in case we need this for the API call)
len(df.index)
```

```python
state = []
for x in range(len(df.index)):
    state.append('CA')
```

```python
df['State'] = state
df.head()
```

```python
df = df.rename(columns={'Region': 'City'})

```

```python
df.to_csv('cleaned_city_data.csv')
```
