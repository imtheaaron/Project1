```python
#grab the temp data for each month and add to lists that we will append to our dataframe

cities = df.City
jan = []
feb = []
mar = []
apr = []
may = []
jun = []
jul = []
aug = []
sep = []
oct = []
nov = []
dec = []

for city in cities:
    url = 'http://api.worldweatheronline.com/premium/v1/weather.ashx?key=42861775afa64821bac83523181404&q='+city+',CA&format=json'
    response = requests.get(url).json()
    jan.append(response['data']['ClimateAverages'][0]['month'][0]['absMaxTemp'])
    feb.append(response['data']['ClimateAverages'][0]['month'][1]['absMaxTemp'])
    mar.append(response['data']['ClimateAverages'][0]['month'][2]['absMaxTemp'])
    apr.append(response['data']['ClimateAverages'][0]['month'][3]['absMaxTemp'])
    may.append(response['data']['ClimateAverages'][0]['month'][4]['absMaxTemp'])
    jun.append(response['data']['ClimateAverages'][0]['month'][5]['absMaxTemp'])
    jul.append(response['data']['ClimateAverages'][0]['month'][6]['absMaxTemp'])
    aug.append(response['data']['ClimateAverages'][0]['month'][7]['absMaxTemp'])
    sep.append(response['data']['ClimateAverages'][0]['month'][8]['absMaxTemp'])
    oct.append(response['data']['ClimateAverages'][0]['month'][9]['absMaxTemp'])
    nov.append(response['data']['ClimateAverages'][0]['month'][10]['absMaxTemp'])
    dec.append(response['data']['ClimateAverages'][0]['month'][11]['absMaxTemp'])
```

```python
#the temp data looks like a float. We'll need to sum this data, so check the type.
type(jan[0])
```

```python
#add the montly temperature data to our dataframe

df['jantemp'] = jan
df['febtemp'] = feb
df['martemp'] = mar
df['aprtemp'] = apr
df['maytemp'] = may
df['juntemp'] = jun
df['jultemp'] = jul
df['augtemp'] = aug
df['septemp'] = sep
df['octtemp'] = oct
df['novtemp'] = nov
df['dectemp'] = dec
df.head()
```

```python
#change the temp data to floats

df.jantemp = pd.to_numeric(df.jantemp)
df.febtemp = pd.to_numeric(df.febtemp)
df.martemp = pd.to_numeric(df.martemp)
df.aprtemp = pd.to_numeric(df.aprtemp)
df.maytemp = pd.to_numeric(df.maytemp)
df.juntemp = pd.to_numeric(df.juntemp)
df.jultemp = pd.to_numeric(df.jultemp)
df.augtemp = pd.to_numeric(df.augtemp)
df.septemp = pd.to_numeric(df.septemp)
df.octtemp = pd.to_numeric(df.octtemp)
df.novtemp = pd.to_numeric(df.novtemp)
df.dectemp = pd.to_numeric(df.dectemp)
```

```python
#sum each city's temp data in a new column called 'MaxTempSum'

df['MaxTempSum'] = (df.jantemp+df.febtemp+df.martemp+df.aprtemp+df.maytemp+df.juntemp+df.jultemp+df.augtemp+df.septemp+df.octtemp+df.novtemp+df.dectemp)
df.head()
```
